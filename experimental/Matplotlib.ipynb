```python
import matplotlib.pyplot as plt # this is the main canvas, looks like subplots too?
```

```python
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r-') # can do matlib style stuff to the graph
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
# more than 1 plot on a single canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g')
```

```python
# same as the previous example, preferred method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('The Title')
```

```python
axes # interesting - it is just an array of axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # since it is an array, we can iterate through it

for current_ax in axes:
        current_ax.plot(x, y) 
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y) # you can access the subplot axes object | statements need to be on combined lines or it just generates info
```

```python
# figure size and DPI
```

```python
fig = plt.figure(figsize = (10, 2)) # figsize = xx, dpi = 100

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(figsize = (10, 2)) # figsize = xx, dpi = 100

axes.plot(x, y)
```

```python
# legends

fig1 = plt.figure()

ax1 = fig1.add_axes([0, 0, 1, 1])
ax1.plot(x, x**2, label = 'x sqrd')
ax1.plot(x, x**3, label = 'x cubed')

ax1.legend() # has a loc = code (0) is the best, tons of others, or tuple for lower left corner
```
