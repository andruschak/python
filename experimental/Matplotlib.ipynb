```python
import matplotlib.pyplot as plt # this is the main canvas, looks like subplots too?
```

```python
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r-') # can do matlib style stuff to the graph
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
# more than 1 plot on a single canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
```

```python
# same as the previous example, preferred method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('The Title')
```
