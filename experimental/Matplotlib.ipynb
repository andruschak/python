```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r-') # can do matlib style stuff to the graph
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
# more than 1 plot on a single canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g')
```
