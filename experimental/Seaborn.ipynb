```python

# Advanced plotting library
# https://github.com/mwaskom/seaborn
# http://seaborn.pydata.org/examples/index.html

# conda install seaborn
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # seaborn has some built in data sets
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill']) # looks like most tips were between 10 - 20$ and taper off
```

```python
sns.distplot(tips['total_bill'], kde=False) # Kernel Density Estimation
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30) # higher resolution - find the sweet spot
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips) # comparing to columns - high higher the bill the higher the tip
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') # fade instead of every plot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') # regression line
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
tips.head()
```

```python
sns.pairplot(tips) # pass in the data frame and it will do comparisons on all numerical columns
```

```python
sns.pairplot(tips, hue='smoker') # pass in categorical columns (smoker in this instance is either y or n)
```

```python
sns.pairplot(tips, hue='smoker', palette='coolwarm')
```
